```python
import csv

class Persona:
    
    # Clase constructor que recibe el nombre y la edad de la persona.
    def __init__(self):
        self.nombre = input("Ingrese el nombre de la persona: ")
        self.edad = int(input("Ingrese la edad de la persona: "))
        
    def set_nombre(self):
        pass

    def get_nombre(self, nombre):
        self.nombre = nombre

    def get_edad(self, edad):
        self.edad = edad
    
    def print_persona(self):
        print(f"{self.nombre} tiene {self.edad} años.")
    
    def es_mayor_de_edad(self):
        mayorDeEdad = self.edad > 18
        return mayorDeEdad

    # Método estático que compara la edad entre el objeto nuevo y el anterior.
    @staticmethod
    def es_mayor_que(esMayorQue, edad, nombre):
        print(f"Edad del objeto actual: {esMayorQue}")
        print(f"Edad de {nombre}: {edad}")
        
        if esMayorQue > edad:
            print(f"El objeto actual es mayor que {nombre}")
        elif esMayorQue < edad:
            esMayorQue = edad
            print(f"{nombre} es mayor que el objeto actual.")
        elif esMayorQue == edad:
            print(f"El objeto actual y {nombre} tienen la misma edad.")

        return esMayorQue

    # Método estático que recibe dos objetos Persona y compara sus edades.
    @staticmethod
    def get_mayor(nombre1, edad1, nombre2, edad2): 
        if edad1 > edad2:
            print(f"{nombre1} es mayor que {nombre2}.")
        elif edad1 < edad2:
            print(f"{nombre2} es mayor que {nombre1}.")
        elif edad1 == edad2:
            print(f"{nombre1} y {nombre2} tienen la misma edad.")
    
    # Método estático que recibe el nombre del archivo y la lista de personas. Este crea un nuevo archivo.
    @staticmethod
    def dump_csv(nombreArchivo, listaPersonas):
        nombreArchivo = f"{nombreArchivo}.csv"
        columnas = [ "Nombre", "Edad" ]
        filas = listaPersonas

        with open(nombreArchivo, 'w', newline='') as csvfile: # El argumento newline='' evita que se generen filas en blanco en el archivo.
            # Crear un nuevo objeto con csv writer 
            csvwriter = csv.writer(csvfile) 
            
            # Escribir columna en el archivo nuevo
            csvwriter.writerow(columnas) 
            
            # Escribir las filas en el archivo nuevo
            csvwriter.writerows(filas)
    
    # Método estático que recibe el nombre del archivo. Este abre el archivo e imprime su contenido.
    @staticmethod
    def load_csv(nombreArchivo):
        nombreArchivo = f"{nombreArchivo}.csv"
        filas = listaPersonas

        with open(nombreArchivo, 'rt') as csvfile: 
            # Crear un nuevo objeto con csv reader 
            csv_rows = csv.reader(csvfile)
            
            # Iterar el objeto e imprimir su contenido
            for fila in csv_rows:
                print(fila)


esMayorQue = 0 # Declarar edad de elemento original.

print("-------------------- Primer persona --------------------")


persona1 = Persona()
persona1.print_persona()
print(f"{persona1.nombre} es mayor de edad? {persona1.es_mayor_de_edad()}")
esMayorQue = persona1.es_mayor_que(esMayorQue, persona1.edad, persona1.nombre)


print("-------------------- Segunda persona --------------------")


persona2 = Persona()
persona2.print_persona()
print(f"{persona2.nombre} es mayor de edad? {persona2.es_mayor_de_edad()}")
esMayorQue = persona2.es_mayor_que(esMayorQue, persona2.edad, persona2.nombre)


print("-------------------- ¿Quién es mayor? --------------------")


persona2.get_mayor(persona1.nombre, persona1.edad, persona2.nombre, persona2.edad)


print("-------------------- Manejo de archivos --------------------")

listaPersonas = [
    [ persona1.nombre, persona1.edad ],
    [ persona2.nombre, persona2.edad ],
]

persona2.dump_csv("Tarea 4 - Personas", listaPersonas)
print("Se ha creado el archivo Tarea 4 - Personas satisfactoriamente.")
print("Esta es la información del archivo Tarea 4:")
persona2.load_csv("Tarea 4 - Personas")
```
